# Visualização dos dados do dataset 'Heart Disease' da base de Cleveland

```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.set(style="white", color_codes=True)
```

```python
colunas = ['age', 'sex', 'pain_type', 'blood_pressure',
           'serum', 'fasting_blood_sugar', 'restecg', 'max_hrate',
           'exang', 'oldpeak', 'slope', 'ca', 'thal', 'num']
full_data = pd.read_csv("src/processed.cleveland.data.csv", names=colunas, na_values=["?"])

full_data.head()
```

```python
full_data['num'].value_counts()
```

```python
# Limitar a visualização de dados em presença de doença do miocárdio (1) ou ausêcia (0)
full_data.loc[full_data['num'] > 0, 'num'] = 1

# Estratégia pra lidar com dados faltantes: dispensar a linha inteira (são poucas linhas)
full_data = full_data.dropna()
```

```python
full_data['num'].value_counts()
```

### Plot de correlações para ver se tem algum par de features que se comportam parecido

```python
# Compute the correlation matrix
corr = full_data.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

Podemos perceber no diagrama acima que as features que mais correlacionam são max_hrate, exang, oldpeak, slope, ca, e thal.

### Pairplot de max_hrate, exang, oldpeak, slope, ca, e thal

```python
pairplot = ['max_hrate', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'num']

pairplot_data = full_data[pairplot]

sns.pairplot(pairplot_data, hue="num")
```

### Gráfico combinando dispersão bivariada e histogramas univariados

```python
for plot in pairplot:
    for plot2 in pairplot:
        sns.jointplot(x=plot2, y=plot, data=full_data, size=5)
    plt.show()
```

```python
for plot in pairplot:
    fig, ax = plt.subplots(ncols=len(pairplot)-1, sharey=True, figsize=(20, 5))
    for i, plot2 in enumerate(pairplot[:-1]):
        sns.stripplot(x=plot2, y=plot, data=full_data, jitter=0.4, hue="num", ax=ax[i])
    fig.show()
```
