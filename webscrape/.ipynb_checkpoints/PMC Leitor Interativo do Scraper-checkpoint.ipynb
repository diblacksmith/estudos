### Setup

```python
%matplotlib inline
import webbrowser
from ast import literal_eval
import re
from datetime import datetime
from time import sleep
from IPython.display import clear_output
import pandas as pd
from matplotlib import pyplot as plt


filename = "stems/final.txt"
with open(filename,'r') as myf:
    print '[*] Extraindo artigos de', filename
    s = myf.read()
    stems = literal_eval(s)
```

### Quantidade de artigos por termo de pesquisa

```python
print 'index','termo'.rjust(35),'quantidade'.rjust(32)
soma = 0
for i,key in enumerate(stems.keys()):
    soma += len(stems[key])
    print str(i).ljust(3), key.rjust(60),len(stems[key])
print str(soma).rjust(66)
```

### Seleção interativa

```python
print 'index','termo'.rjust(35),'quantidade'.rjust(32)
for i,key in enumerate(stems.keys()):
    print str(i).ljust(3), key.rjust(60),len(stems[key])
    
index = int(raw_input("Seleção: "))


for e,key in enumerate(stems.keys()):
    if e < index:
        continue
    i = 0
    sair_flag = False
    while i < len(stems[key]):
        clear_output(wait=True)
        print key,' = ',len(stems[key])
        print '>>>>>>>>>', key
        article = stems[key][i]
        print article['title'],'\n'
        print "Publicação:",article['pub_date']
        print article['kwd'],'\n'
        print article['p'],'\n'
        
        apagar = re.compile("(apa|del|rem)",re.IGNORECASE)
        url = re.compile("(url|link|ender|site|abr|ope)",re.IGNORECASE)
        proximo = re.compile("(con|ava|pro|nex)",re.IGNORECASE)
        sair = re.compile("(sai|lea|exi|fin)",re.IGNORECASE)
        
        match = False
        while not match:
            sleep(0.5)
            act = raw_input("Ação ")
            if apagar.match(act):
                print 'apagando'
                stems[key].remove(article)
                match = True
                i -= 1
            elif url.match(act):
                print 'abrindo'
                i -= 1
                sleep(0.5)
                webbrowser.open(article['url'])
                match = True
                sleep(1.5)
            elif proximo.match(act) or act == '':
                print 'continuando'
                match = True
                continue
            elif sair.match(act):
                print 'saindo'
                i = len(stems[key])
                match = True
                sair_flag = True
            else:
                print 'invalido!'
        i += 1
    filename = "stems/thread_stems" + str(datetime.now()).replace(' ','-') + '.txt'
    with open(filename, 'w') as myf:
        myf.write(str(stems))
        print '[*] resultados salvos em',filename
        
    if sair_flag:
        break
```

### Salvando

```python
filename = "stems/thread_stems" + str(datetime.now()).replace(' ','-') + '.txt'
with open(filename, 'w') as myf:
    myf.write(str(stems))
    print '[*] resultados salvos em',filename
```

## Visualizando os números

### Gerais

```python
data = [[],[],[], []]
for key,value in stems.iteritems():
    data[0].append(len(value))
    split = key.split(' AND ')
    data[1].append(split[0])
    data[2].append(split[1])
    data[3].append(key)
    
df = pd.DataFrame(data={'quantidade':data[0],
                        'primary':data[1],
                        'secondary':data[2],
                        'full':data[3]}
                 )
# Ordenar pela quantidade
df = df.sort_values(by='quantidade', ascending=False)
# Resetar o índice sem manter o velho como uma antiga coluna
df = df.reset_index(drop=True)
# Escolher a ordem que as colunas irão aparecer
df[['primary','secondary','quantidade']].head(10)
```

```python
# Tamanho da figura
plt.figure(figsize=(9,9))

# Setar linhas do grid
ax = plt.axes()        
ax.xaxis.grid(True)
tam = len(df['quantidade'])
plt.barh(range(1,tam + 1), [df['quantidade'][x] for x in range(tam)], alpha=0.65, zorder=3)

# Setar texto e ticks
plt.yticks(range(1,tam + 1),df['full'])
plt.xticks(range(12))
plt.xlabel('Quantidade de artigos')
plt.title(u'Número de artigos por termo de pesquisa')
```

### Específicos

```python
primary = df.groupby(['primary']).sum()
secondary = df.groupby(['secondary']).sum()

# Ordenar pela quantidade
primary = primary.sort_values(by='quantidade', ascending=False)

# Ordenar pela quantidade
secondary = secondary.sort_values(by='quantidade', ascending=False)
```

## Jogando no formato da planilha

```python
data_ordenada = []

### Transformar '2016 Jan' em 2016-01-01
for key in stems:
    for artigo in stems[key]:
        artigo['pub_date'] = datetime.strptime(artigo['pub_date']+" 01", "%Y %b %d")
```

```python
### Ordenar os artigos por data    
for key in stems:
    data_ordenada.extend(stems[key])
    
data_ordenada.sort(key = lambda a: a['pub_date'])
```

```python
### Jogar num .txt pra ficar fácil de copiar e colar na spreadsheet

with open('planilha.txt', 'w') as myf:
    for artigo in data_ordenada:
        title = artigo['title'].replace('\n','')
        hyperlink = '=HYPERLINK("%s";"%s")' % (artigo['url'], title)
#       myf.write(hyperlink.encode('utf8') + '\n')
        myf.write(artigo['pub_date'].strftime("%Y %b") + '\n')
```
